# Advanced Operations

```python

df['squared age']=df['age'].apply(lambda x:x**2)
df
```

```python
#replace
df['gender'].replace({'Male':1,'Female':0},inplace=True)
df
```

```python

def calculate_tax(salary):
    if salary>5000:
        return salary*0.05
    else:
        return 0
df['tax']=df['Base Salary'].apply(calculate_tax)
df
```

```python
salary_range=df['Base Salary'].max()-df['Base Salary'].min()
salary_range
```

```python
print('mean before removing outliers is ',df['Base Salary'].mean())
Q1=df['Base Salary'].quantile(0.25)
Q3=df['Base Salary'].quantile(0.75)
IQR=Q3-Q1
print('\nIQR is ',IQR)
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print('\nlower bound is ',lower_bound)
print('\nupper bound is ',upper_bound)
df=df[(df['Base Salary']>=lower_bound) & (df['Base Salary']<=upper_bound)]
df
```
